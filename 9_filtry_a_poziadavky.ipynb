Загрузим нашу таблицу, и взглянем на неё ещё раз.

```python
import pandas as pd

data = pd.read_csv("D:\\new_data\\Metro_Interstate_Traffic_Volume.csv")
data.head(3)
```

Что если нам необхдимо отобрать все колонки, которые содержат какой-то символ или набор символов. Например, нам нужно отобрать 
все колонки, которые содержат слово "weather". Конечно, мы можем воспользоваться метотдами loc или iloc и просто записать 
назавание или порядковый номер каждой колонки. Но, во первых, это не очень удобно, а во-вторых, если наша таблица будет 
содержать 1000 колонок, то в ручную выбирать из тысячи нужные нам - удовольствие ниже стреднего.
В таких случаях на помощь к нам придёт метод filter(), у которого есть параметр "like", который на вход может принимать набор символов и отфильтровывать по ним наши данные. Давайте выбирем все колонки, в название которых входит слово "weather".

```python
data.filter(like="weather")
```

Также мы можем отфильтровать данные по индексу. Например, мы хотим получить все строки, в которых содержиться значение 1111. Для того, чтобы работать с индексом необохдимо задать параметр axis=0. 

```python
data.filter(like='1111', axis=0)
```

Фильтрация по индексу бывает полезна, например, когда в качестве индекса у нас троковые значения. Давайте установим колонку 
"weather_main" как индекс и отберём только те значения, где индекс = 'Snow'.

```python
data_new = data.set_index('weather_main')
data_new
```

```python
data_new.filter(like='Snow', axis=0)
```

Мы также можем выбрать конкретные колонки (тоже самое может сделать с помощью метода loc).

```python
data.filter(items=["rain_1h", 'clouds_all', 'date_time'])
```

Пару небольших трюков, не относящихся к методу filter(), но полезные при фильтрации данных. Как быстро получить список всех колонок DataFrame? Просто нужно обернуть наш DataFrame в список.

```python
columns = list(data)
columns
```

Допустим, нам нужно вернуть список всех колонок, которые содержат определённый набор символов. Можно, конечно, написать обычный цикл и перебрать все значения. Но проще написать свой генератор последовательности.

```python
weather_columns = [column for column in list(data) if 'weather' in column]
weather_columns
```

Что мы тут сделали? Создали список с названием всех колонок при помощи комманды list(data) и выбрали из него те колонки, в название которых входит слово "weather".

Теперь давайте посмотрим на метод query(). Этот метод посзволяет более просто и лаконично отбирать данные по условиям. Всё, что нам нужно передать в качестве аттрибута - это наш запрос. Двайте, нарпимер, отберём только те строки, где значение колонки 
"Clouds_all < 40".

```python
data.query("clouds_all  < 40")
```

Обратите внимание, что целый запрос мы должны написать в ковычки, но название колонки пишем без ковычке. В метод query() можем предать несколько условий. Каждое условие необоходимо отделить круглыми скобками.

```python
data.query("(traffic_volume < 500) & (weather_description == 'broken clouds')")
```

У метода query() есть ещё одна интересная особенность, он может принимать в качестве аргумента ссылку на внешний объект. Давайте посмотрим, что это значит. Создадим две переменные podmienka_1 и podmienka_2 и сохраним в них значенийя условий из предыдущего примера. Затем, подставим подствим ссылку на эти переменные в запрос. Чтобы метод query() понял, что это внешняя ссылка необходимо перед переменной поставить знак "@".

```python
podmienka_1 = 500
podmienka_2 = 'broken clouds'

data.query("(traffic_volume < @podmienka_1) & (weather_description == @podmienka_2)")
```

Такая запись улучшать читаемость вашего кода. Кроме того, это если вы используете одинаковые условия в разных запросах, то вам не нужно будет менять условия в каждом запросе. Достаточно изменить значение переменной.
